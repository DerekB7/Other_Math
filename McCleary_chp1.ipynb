**ex 1.1**  
Let $f:A \longrightarrow B$ be any function with $U,V$ subset of $B$ and $X$ a subset of $A$.  Prove the following about the about the pre-image operation.  

(**note:** $f^{-1}$ is the pre-image operation from the bottom of page 2 which is not a function and $f^{-1}$ *is not* the *function* $f^{-1}$ defined on a power set given on top of page 3 which is an unfortunate overloading of notation)  

**also note:** $\subset$ here should be interpreted as $\subseteq$ **not** *as proper subset*  

**(a)** $U\subset V  \implies f^{-1}(U) \subset f^{-1}(V)$  
note: ignoring the role of empty sets, this is pictorially obvious if we sketch out and shade circles for these sets 

(i) 
suppose $u \in U$ has no preimage under $f$.  Then $f^{-1}(u) \in f^{-1}(U)$ by construction 
and $f^{-1}(u) \in f^{-1}(V)$ as this constains the empty set as well   
*(note: treatment of empty sets should be cleaned up though this is worthwhile discussion point)*  

(ii)  
suppose $u \in U$ has a preimage under $f$, then  
$a = f^{-1}(u)$  
and $f^{-1}(u) \in f^{-1}(U)$ by construction and $f^{-1}(u)=a \in f^{-1}(V)$  because $u\in V$  

**(b)** $f^{-1}(U\cup V) = f^{-1}(U)\cup f^{-1}( V)$  
again this is pictorially obvious  
for $u \in U$  
$f^{-1}(u) \in f^{-1}(U\cup V)$ by construction and   $f^{-1}(u) \in f^{-1}(U)$ by construction  
the argument is identical for $v$  


**(c)** $f^{-1}(U\cap V) = f^{-1}(U)\cap f^{-1}( V)$  
this is nearly implied by (a) since $U\cap V \subset U$ and $U\cap V \subset V$ then  
$f^{-1}(U\cap V)\subset f^{-1}(U)\cap  f^{-1}(V)$  
a better approach is to use (b)  
(i) supppose $U\cap V = \{\}$  then  
$f^{-1}(U\cap V) = \{\}= f^{-1}(U)\cap  f^{-1}(V)$   

(ii) if they are not disjoint, then  biparition $U$ and $V$ such that $W := U \cap V$  and $U= U' \cup W$ (i.e. $U':=U-W$) and $V=V'\cup W$ (i.e. $V':=V-W$) which implies $U' \cap V' = \{\}$  

Then  
$f^{-1}\big(U\cap V\big)$  
$=f^{-1}\big((U'\cup W) \cap (V'\cup W)\big)$  
$=f^{-1}\big((U'\cap V') \cup (U' \cap W) \cup (W\cap V') \cup (W\cap W)\big)$   
$=f^{-1}\big(\{\} \cup \{\} \cup \{\} \cup W\big)$   
$=f^{-1}\big(W\big)$   
$=f^{-1}\big(W\big)\cap f^{-1}\big(W\big)$   
$=\Big(f^{-1}\big(W\big)\cup f^{-1}\big(U'\big)\Big)\cap f^{-1}\big(W\big)$      
$=\Big(f^{-1}\big(W\big)\cup f^{-1}\big(U'\big)\Big)\cap \Big(f^{-1}\big(W\big)\cup f^{-1}\big(V'\big)\Big)$      
$=f^{-1}\big(W\cup U'\big)\cap f^{-1}\big(W\cup V'\big)$  (by b)      
$=f^{-1}\big(U)\cap f^{-1}\big(V\big)$   

**(d)** $f\big(f^{-1}(U)\big)\subset U$   
if $f^{-1}(U)=\{\}$, the result is immediate.  Otherwise write $U = U' \cup U''$   
where $U'$ contains all values with a non-empty pre-image under $f$ and $U'' = U-U'$.  Then    

$f^{-1}(U)$  
$=f^{-1}(U'\cup U'')$  
$=f^{-1}(U')\cup f^{-1}(U'')$ (by b)   
$=f^{-1}(U')\cup \{\}$  
$=f^{-1}(U')$   
and  $f\big(f^{-1}(U')\big) =U$  (i.e. select arbitrary $u' \in U'$ and for arbitrary $a\in f^{-1}(u')$ we have $f(a) = u'$)  \
thus  
$f\big(f^{-1}(U)\big)=f\big(f^{-1}(U')\big)= U'\subset \big(U'\cup U''\big) = U$   

**(e)** for $X\subset A$ then $X\subset f^{-1}\big(f(X)\big)$  
this is immediate: for $x\in X$, $f(x) = y$ and then $y$ under the pre-image of $f$ by definition includes $x$.  Note this may be a proper subset.  In particular consider the restriction of $f$ to $X$ given by $f\vert_{ X}$.  

$X= f\vert_{ X}^{-1}\big(f\vert_{ X}(X)\big)$ and if $f$ is injective, then $f\vert_{ X}^{-1}\big(f\vert_{ X}(X)\big)=f^{-1}\big(f(X)\big)$  but if $f$ is not injective this need not be the case  


**(f)** and **(g)** are if for any $U\subset B$, $f\big(f^{-1}(U)\big)=U$ then $f$ is surjective and if for any $X\subset A$, $f^{-1}\big(f(X)\big)=X$ then $f$ is injective, respectively.   


note: the use of 'any' should be interpreted as 'arbitrary' i.e. in same manner as in analysis where we say "for any $\epsilon \gt 0$ there exists $\delta \gt 0$..."


The results are nearly immediate: for the first case simply select $U:=B$ and by the argument in (d) this implies $U''$ has an empty pre-image under $f$ and hence it is empty thus $f$ is surjective.  The second case follows similarly, except let $X$ be a set containing an arbitrary single point $\in A$, then $f^{-1}\big(f(X)\big)=X$ is the definition of injective-- i.e. every point in the image of $f$ has exactly one point in its pre-image (since the choice of point in $A$ was arbitrary).
