$K\subseteq \bigcup_{\alpha \in J}C_\alpha$ where all $C_\alpha$ are open  

and since these are open sets and we have a countable basis  
$C_\alpha = \bigcup_{k\in S_{\mathbb N_\alpha}}B_k$  
where $S_{\mathbb N_\alpha}$ is a subset containing a subset of the natural numbers  

$K\subseteq \bigcup_{\alpha \in J}C_\alpha = \bigcup_{\alpha \in J}\Big(\bigcup_{k\in S_{\mathbb N_\alpha}}B_k\Big)  \subseteq \bigcup_{k=1}^\infty B_k= X$  

thus  
$\bigcup_{\alpha \in J}\Big(\bigcup_{k\in S_{\mathbb N_\alpha}}B_k\Big)=\bigcup_{i\in S}B_i $  
for some $S\subseteq \mathbb N$   

And to each $B_i$ we may assign one $C_\alpha$ such that  
$B_i \subseteq \bigcup_{k\in S_{\mathbb N_\alpha}}B_k= C_\alpha $  
i.e. where $i\in S_{\mathbb N_\alpha}$  
at the cost of some notational overload, we relabel such $C^{'}_i:=C_\alpha$  
note: there is no need for uniqueness between $C^{'}_i$ and $C^{'}_k$   

taking unions of each side we get  
$K\subseteq \bigcup_{i\in S}B_i \subseteq \bigcup_{i\in S}\bigcup_{k\in S_{\mathbb N_\alpha}}B_k= \bigcup_{i\in S}C^{'}_i$  
where the RHS is a countable union i.e. a countable subcover as required
