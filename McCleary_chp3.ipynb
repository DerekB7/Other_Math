**ex 3.4**  

A topological space $X$ is called a **metrizable space** if the topology on $X$ can be induced by a metric space structure on $X$.  Not every topology on a set comes about in this fashion [e.g. metrics imply unique limits and we saw examples in the chapter that this isn't generally true for non-Hausdorff spaces].   

Show that a metric space is always (i) Hausdorff and (ii) 1st countable.   

*(i)* metric space $\implies$ Hausdorff  
referencing def 3.8 (p. 33), for two (distinct) arbitrary  $x,y\in X$ , where $x\neq y$, using our metric we have  
$d(x,y)=3\delta \gt 0$  
Then $U:=B\big(x,\delta\big)$ and $V:=B\big(y, \delta\big)$ are two open sets with $x\in U$ and $y\in V$ and  
$U\cap V = \big\{\big\}$  
hence this space is Hausdorff  

(Formally:  suppose $z \in U\cap V$. Then $d(x,z) \lt \delta$ and   $d(z,y) \lt \delta$ but by triangle inequality we have  $3 \delta = d(x,y)  \leq d(x,z) + d(z,y) \lt 2 \delta$ which is a contradiction.)   



*(ii)* metric space $\implies$ 1st countable  
referencing def 3.13 on p.35  we need to show for arbitrary $x\in X$ there is a collection of sets $\big\{U_k^{(x)}\vert k=1,2,3..\big\}$ such that for any open $V$ in $X$ containing $x$ (i.e. $x \in V$, there is one of these open sets with $U_j^{(x)}$ such that $x \in U_j^{(x)}\subseteq V$)     

define $U_k^{(x)}:= B\Big(x, \big(\frac{1}{2}\big)^k\Big)$   
referencing example 3 on page 23, any metric space has a basis given by a collection $B(x', \epsilon)$, so $V$ (by prop 2.7 on page 22) is generated by unions of these open balls and $x$ must be in one of said balls  
(if not, then   $B_\alpha\subseteq X-\{x\}$ for all $\alpha$ in the union, thus  $V = \bigcap_\alpha B_\alpha\subseteq \bigcap_\alpha X-\{x\}=X-\{x\}$ and $V$ would not contain $\alpha$)  
so let $B_{\alpha^*}= (x^*, \epsilon)$ contain $x$.  Thus $0\lt d(x,x^*)= \eta \lt \epsilon$  and set $m:= \epsilon -\eta\gt 0$  

since the sequence $\big(\frac{1}{2}\big)^k\to 0$ (monotone decreasing, bounded below by zero...),  
there exists some $K$ such that for all $k\geq K$:   
$\big(\frac{1}{2}\big)^k\leq m$  

then setting $j:= K$ (or for a nicer picture $j:= K+1$), we have found a satisfying  
$x \in U_j^{(x)}\subseteq V$   
and hence the metric space is first countable

**ex 3.6**  

Suppose that $(X,d)$ is a metric space and $A\subseteq X$.  Define the distance from $A$ to a point $x$ as  
$d(\_,A): X\longrightarrow \mathbb R$ given by $d(x,A) = \inf_{a\in A}d(x,a)$  

**(i.)** Show that $d(\_,A): X\longrightarrow \mathbb R$ is a continuous function   
- - - -    
*commentary about the image of* $d(\_,A)$   
first consider 'flipping' things around applying the metric, with fixed $x$ to all $a\in A$  
$d^{(x)}: A\longrightarrow \mathbb R_{\geq 0}$  
This gives us a collection of intervals on the real half line.  Since this is bounded below by zero, a lower bound exists hence an infimum, $L$ exists. If $L \in \text{image }d^{(x)}$ then $L$ is the minimum and we are done.  If $L \not\in \text{image }d^{(x)}$ then it lives in some interval $(L,m]$ or $(L,m)$.  

This says that the image of $A$ has points arbitrarily close to $L$ and implies that we may select from the pre-image some $a' \in A$ such that $d(x,a')-d(x,A)\lt \epsilon'$ for arbitrary choice of $\epsilon' \gt 0$.  We exploit this in the proof of continuity of $d(\_,A)$ that follows.        
- - - -    

Since metric spaces are 1st countable this is precisely what Corollary 3.17 (page 37) is getting at with its converse -- i.e. the converse says that if $f$ is not continuous (in a 1st countable space) then there exists some convergence sequence where $x_k\to x$ but $f(x_k)\not \to f(x)$.  
We suppose for contradiction that $d(\_,A)$ is not continuous and that we have identified said sequence which must exist. Thus there is some (fixed) $\delta \gt 0$ such that  

$ \delta \leq \big \vert d^*(x_k, A) -d^*(x, A)\big \vert $  for all $k$. And from here the result follows by reverse triangle inequality.   

(a) if $d^*(x_k, A) -d^*(x, A)\geq 0$  
$ \delta \leq  d^*(x_k, A) -d^*(x, A)\leq \big(d(x_k,a') + \epsilon^{'}\big) -\big(d(x,a^*)-\epsilon^*\big)\leq d(x_k,a^*)-d(x,a^*)+ \epsilon^{'}+\epsilon^*\leq d(x_k,x)+ \epsilon^{'}+\epsilon^*$   
note: the above assumes WLOG that $d(x_k,a') \leq d(x_k,a^*)$ -- if this was not true we'd instead select $a':= a^*$ to make it true  

and $d^*(x_k, A) -d^*(x, A)\lt 0$ uses a virtually identical argument to get the same upper bound, thus we have  

$\delta \leq  d(x_k,x)+\epsilon^{'}+\epsilon^*$   
but $d(x_k,x) \lt \epsilon$ for $k$ large enough and by selecting $\epsilon:= \epsilon^{'}:= \epsilon^{*}:=\frac{1}{4}\delta$ this   implies  
$\delta \leq  d(x_k,x)+\epsilon' +\epsilon^*\lt \epsilon + \epsilon'+\epsilon^* = \frac{3}{4}\delta$   
which is a contradiction  



**(ii.)**  Show that a point $x \in X$ is in the closure of $A$ *iff* $d(x,A)=0$    

*direction 1:*  
$x \in \text{cls }A\implies d(x,A)=0$  
note: if $x\in A$ the result is immediate by selecting $x:=a$.  
per ex 3.4 we know this metric space is 1st countable.  So we use Prop 3.14 (page 36) and see that $x$ is the limit of a sequence of $x_k \in A$  
thus $x_1,x_2, x_3,... \to x\implies d(x_1,A),d(x_2,A), d(x_3,A),... \to d(x,A)$  
by Corollary 3.17 (p. 37 or  standard analysis arguments)  and the sequence on the right is identically 0, thus the limit is zero as well.  

*direction 2:*  
$d(x,A)= 0 \implies x \in \text{cls }A $  


re-visiting "commentary about the image", we see that if $x \in A$ the result is immediate.  If not, we have 
$d(x,a') =d(x,a')-d(x,A)\lt \epsilon$ for any $\epsilon \gt 0$   
which implies a sequence  
$d(x,a_k) =d(x,a_l)-d(x,A)\lt \big(\frac{1}{2}\big)^k$

i.e. since $d$ is a metric, we see $\lim_{k\to \infty} d(x,a_k) = 0 \implies \lim_{k\to \infty} a_k = x$  
and since the limits are unique in metric (and more generally Hausdorff) spaces, this is an equality (not a mere equivalence relation)        


Thus 
$x =\lim_{k\to \infty} a_k\in \text{cls }A$ 
(ref p. 29-31, in particular Prop 3.2 or typical analysis arguments).

**ex 3.7**   
Prove that the following are topological properties  
*(1)* $X$ is a separable space   
Let $f:X\longrightarrow Y$ be a homeomorphism and $A\subseteq X$ be countable and dense  (i.e $\text{cls }A = X$) then  using Theorem 3.16(3) 

$Y= f\big(X)= f\big(\text{cls }A\big)\subseteq \text{cls }f\big(A\big)=\text{cls } B \subseteq Y$  
where $B=f(A)$ is countable (invertible maps preserve cardinality) and $B=Y$ hence $Y$ is separable  

*(2)* $X$ satisfies the Hausdorff condition  
Let $f:X\longrightarrow Y$ be a homeomorphism with $X$ Hausdorff, which tells us that for $x\neq x'$ there exists open sets $U_X, V_X$  with $x\in U_X$, $y \in V_X$ and $U\cap V = \big\{\big\}$  

then  
$y=f(x)$, $y\in U_Y= f\big(U_X\big )$    
$y'=f(x')$, $y'\in V_Y= f\big(V_X\big)$     
and $U_Y\cap V_Y = \big\{\big\}$  
(which follows by injectivity of $f$-- i.e. if the intersection was non-trivial and had some $y^*$ then that implies a $x^*$ as the unique pre-image point such that $x^*\in U_x\cap V_x$)  

*(3)*  $X$ has the discrete topology, recalling from page 20 that this means $X$ and all subsets of $X$ are declared open 

Let $f:X\longrightarrow Y$ be a homeomorphism and $g:= f^{-1}$ which is continuous since $f$ is a homeomorphism.    
Select arbitrary $S\subseteq Y$.  Then $g(S) = A$ is open in $X$ because $X$ has the discrete topology.  Applying def 2.9 (p 24), since $A$ is open in the codomain and $g$ is continuous, $g^{-1}(A) = S$ is open in the domain $Y$. Since $S$ was arbitrary, all subsets of $Y$ are open and $Y$ has the discrete topology.
